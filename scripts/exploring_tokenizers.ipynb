```python
import sentencepiece as spm
from pprint import pprint
import matplotlib.pyplot as plt
from constants import ABSOLUTE_PARENT
```

```python
vocab_size_lst = ['1024', '2048', '3000', '3700', '4096', '5000', '6000', '7000', '8192', '9000', '10000',
                  '11000', '12000', '13000', '14000', '15000', '16384', '18000', '20000',
                  '25000', '32768', '65536', '131072']
```

```python
# Directory to save scripts outputs 
working_dir = f"{ABSOLUTE_PARENT}/files/outputs"
# Tokenizer file name
tokenizer_file_name = f"sp_{{0}}_bpe_{{1}}"
# Language
lang = 'hi'
```

```python
tokenizer_path_dict = dict()
for vocab_size in vocab_size_lst:
    # add new pair of key and value
    tokenizer_path_dict.update(
        {f"tokenizer_path_hi_{vocab_size}": f"{working_dir}/sp_{lang}_bpe_{vocab_size}.model"}
    )
pprint(tokenizer_path_dict)
```

```python
def tokenize_text(text, tokenizer_model, vocab_size):
    sp = spm.SentencePieceProcessor()
    sp.Load(tokenizer_model)
    token_pieces_len = len(sp.EncodeAsPieces(text))
    text_char_len = len(text)
    return [vocab_size, text_char_len, token_pieces_len, text_char_len/token_pieces_len]
```

```python
test_text = "कल्पना कीजिए कि आप विदेश में छुट्टी मना रहे हैं, एक ऐसे देश में जहाँ भाषा, मौसम, रहन–सहन, खाना सब कुछ आपके लिए नया है। पहले दिन आप दिन भर घूमते हैं और शाम तक आपको जोर की भूख लग जाती है। आप एक अच्छा खाना चाहते हैं। आपको पता चलता है कि पास में ही एक जगह है जहाँ खाने पीने की कई मशहूर दुकानें हैं। आप उसे आजमाने का फैसला करते हैं।"
```

```python
test_text="विकल्प दो ज्यादा करीब है टेक्निकल एनालिसिस के। यहां पर आप पूरे बाजार में मौके तलाशते हैं यह देखते हुए कि बाजार इस समय किधर जा रहा है और बाजार की पसंद क्या है? टेक्निकल एनालिसिस की तकनीक में बाजार में मौजूद सभी कारोबारियों की पसंद को देखते हुए ट्रेडिंग के मौके ढूंढे जाते हैं। बाजार के ज्यादातर कारोबारियों की पसंद क्या है इस को पहचानने के लिए शेयर या इंडेक्स के चार्ट(chart/graph) को देखा जाता है। कुछ समय बाद उस चार्ट में एक पैटर्न बन जाता है और उस पैटर्न को देखकर आप बाजार का संकेत समझ सकते हैं। टेक्निकल एनालिस्ट (Technical Analyst) का काम है कि वो इस पैटर्न को समझे और अपना नजरिया बनाए। "
```

```python
result = []
for vocab_size in vocab_size_lst:
    result.append(
        tokenize_text(test_text, tokenizer_path_dict[f"tokenizer_path_hi_{vocab_size}"], vocab_size)
    )
pprint(result)
```

```python
# Extract data for the bar chart
tokens = [item[2] for item in result]  # Extract number of tokens
char_per_token = [item[3] for item in result]  # Extract characters per token
```

```python
avg_of_tokens = sum(tokens)/len(tokens)
print(avg_of_tokens)
avg_of_char_per_token = sum(char_per_token)/len(char_per_token)
print(avg_of_char_per_token)
```

```python
# Create the chart
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.plot(range(len(tokens)), tokens, label='Number of Tokens', marker="o")
plt.hlines(avg_of_tokens,0,len(result))
plt.xlabel('Vocab Size')
plt.ylabel('Tokens')
plt.title('Comparison of Number of Tokens and Vocab Size')
plt.xticks(range(len(tokens)), [item[0] for item in result], rotation=45)  # Adjust x-axis labels for better readability
plt.legend()
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.tight_layout()
plt.show()
```
